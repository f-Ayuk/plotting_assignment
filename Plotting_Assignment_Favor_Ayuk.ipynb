```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Data Interpretation:

The visualization tells a story of how vehicle characteristics like engine size and year of manufacture directly influence fuel efficiency and, consequently, CO2 and greenhouse gas emissions across various driving conditions. It also highlights significant redundancies within the dataset, particularly among the different driving cycle metrics and the emission/efficiency scores.

-- pv2 and pv4 are perfectly correlated with themselves (1) but their correlations with other features are generally low

-- id column shows very low correlation values (close to 0) with all other features

-- The year suggests that newer vehicles (year increases) tend to have lower CO2 emissions, better fuel efficiency (feScore), and lower greenhouse gas scores (ghgScore),
